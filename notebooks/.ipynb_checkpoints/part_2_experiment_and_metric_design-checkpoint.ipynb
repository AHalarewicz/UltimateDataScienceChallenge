### 1. What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric.

For Ultimate Inc.'s decision of reimbursing driver partners for the cost of tolls to make sense, the profit produced for Ultimate by its driver partners while operating outside of their home city needs to exceed the the total cost of reimbursed tolls. This means driver partners need to generate enough profit for Ultimate Inc while working in the other city to justify Ultimate Inc.'s decision to cover the cost of tolls. Ultimate Inc. would lose money if the covered the cost of tolls accrued by its driver partners exceed the amount of profits generated by the driver partners while working in the other city. 

Additionally, it is worth considering the supply and demand for drivers in each city. Reimbursements should only be offered at times where the demand for drivers exceeds the supply of available drivers. Otherwise, drivers will steal business from local driver partners. This would negatively impact Ultimate Inc.'s profits as they would have to reimburse tolls to driver partners when a local driver partner could have completed the trip without needing reimbursement. 

If a reimbursement system is being considered, it would be in Ultimate Inc.'s best interest to only reimburse tolls when there is sufficient demand for more drivers while maximizing the amount of trips completed by local drivers. 

A perfect scenario for Ultimate Inc.:

1. Toll reimbursements are only offered when the demand for drivers exceeds the supply in a particular city. 
2. Local drivers have priority when trips are being delegated and assigned to drivers.
3. Driver's from the other city that crossed the bridge must generate enough profit for Ultimate Inc. before returning to their home city and getting reimbursed. 

What metric should be measured? 

Before any concrete toll reimbursement system can be implemented for its driver partners, Ultimate Inc. should consider the mean profits generated by driver partners that complete trips in not their home city. Ideally, Ultimate Inc. should observe the mean profit generated exclusively by driver partners when they are operating outside of their home city.  

### 2. Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. Please provide details on:

###      A. how you will implement the experiment.

To implement this experiment, we would need to select all trips where the driver's home city does not equal the city where the trip began. 

From a table where rows correspond to individual trips, and columns correspond to features describing the rider, driver, and trip details, the query like the following would extract the relevant data:

SELECT driver_id, trip_date, trip_time, sum(trip_fair - drivers_payout) as ultimate_daily_proit

FROM all_trips

WHERE trip_start_city <> driver_home_city 

GROUP BY driver_id, trip_date

The resulting table would consist of rows corresponding to the daily profits from each driver for trips that originated outside of their home city. With this new table, it is now possible to compare daily profits earned to the cost of tolls to be reimbursed. 

The experiment will require two parts:

1. A Hypothesis test to see if the true mean profit generated by drivers working outside of their home city is equal to the fixed cost of tolls (Null Hypothesis). 


2. If the Null Hypothesis is rejected and the true mean is not equal to the cost of tolls, bootstrapping can be used to construct a probability distribution of the expected true mean. Bootstrapping will visualize the difference between the position of the true mean and the expected cost of tolls. 

### B. what statistical test(s) you will conduct to verify the significance of the observation.

Assuming the resulting table contains more than 30 rows of normally distributed data, a z-test would be most appropriate for determining if the mean profit from drivers working outside of their home city is equal to the cost of tolls. The observations collected from the query can be treated as a sample. The z-test will provide a confidence interval that describes the expected range of the true mean profit. If the resulting table from the above query produces less than 30 rows of observations, a t-test would be more appropriate for this experiment.

For this particular experiment, a two sided confidence interval will be most appropriate as it will provide the minimum and maximum of the range expected to contain the true mean.

We can state the null and alternaive hypotheses for the test as the following:

**Null Hypothesis:**

Mean (Daily Profit from Drivers)  ==  Cost of Tolls

**Alternative Hypothesis:**

Mean (Daily Profit from Drivers)  <  Cost of Tolls

OR

Mean (Daily Profit from Drivers)  >  Cost of Tolls


If the resulting p-valus is less than a specified alpha, the Null Hypothis can be rejected. Thus, Indicating the Daily Profit of drivers working outside of their home city does not equal the cost of tolls. 

Bootstrapping can now be used to better visualize if the true mean is greater than the cost of tolls or less than the cost of tolls. 

### C. how you would interpret the results and provide recommendations to the city operations team along with any caveats.

Bootstrapping allows us to simulate the collection of more samples by using the samples that have already been collected. By computing the mean of the bootstrapped distribution of profits, we can observe the expected true mean of profits. This provides the result of the experiment and will provide the most probable true mean profit of drivers operating outside of their home city. As long as the bootstrapped mean is greater than the cost of tolls, it is in Ultimate Inc.'s best interest to reimburse drivers for the cost of tolls. 

Conclusion: 

If the true mean of Ultimate Inc.'s profits generated by drivers working outside of their home city, then Ultimate has the green light implement a Toll Reimbursement system.
