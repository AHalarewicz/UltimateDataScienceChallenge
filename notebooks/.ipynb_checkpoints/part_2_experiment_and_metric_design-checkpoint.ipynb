### 1. What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric.

For Ultimate Inc.'s decision of reimbursing driver partners for the cost of tolls to make sense, the profit produced for Ultimate by its driver partners while operating in the other city needs to exceed the the total cost of reimbursed tolls. This means driver partners need to generate enough profit for Ultimate Inc while working abroad to justify Ultimate Inc.'s reasoning to cover the cost of tolls. Ultimate Inc. would lose money if the covered the cost of tolls accrued by its driver partners exceed the amount of profits generated by the driver partners while working abroad. 

Additionally, its worth considering the supply and demand for drivers in each city. Reimbursements should only be offered at times where the demand for drivers exceeds the supply of available drivers. Otherwise, drivers will steal business from local driver partners. This would negatively impact Ultimate Inc.'s profits as they would have to reimburse tolls to driver partners when a local driver partner could have completed the trip without needing reimbursement. 

If a reimbursement system is being considered, it would be in Ultimate Inc.'s best interest to only reimburse tolls when there is sufficient demand for more drivers while maximizing the amount of trips completed by local drivers. 

A perfect scenario for Ultimate Inc.:

1. Toll reimbursements are only offered when the demand for drivers exceeds the supply in a particular city. 
2. Local drivers have a priority when trips are being delegated and assigned to drivers.
3. Driver's from the other city that crossed the bridge must generate enough profit for Ultimate Inc. before returning to their home city and getting reimbursed. 

What metric should be measured? Before any concrete toll reimbursement system can be implemented for its driver partners, Ultimate Inc. should consider the mean profits generated by driver partners that complete trips in not their home city. Ideally ultimate Inc. should observer the mean profit generated exclusively generated by driver partners when they are operating outside of their home city.
